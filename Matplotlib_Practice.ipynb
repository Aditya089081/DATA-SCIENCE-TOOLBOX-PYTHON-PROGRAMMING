```
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(0,11,2)
y = 2*x+3
plt.plot(x,y,color = 'g',linestyle=":",linewidth = 2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My first plot")
plt.show()
```

```
plt.plot(x,y,color = 'g',linestyle=":",linewidth = 2)
plt.plot(x,5*x,color = "r",linewidth=3,linestyle="--")
plt.title("My first plot")
plt.xlabel("X-axis")
plt.ylabel("Y-label")
plt.grid(True)
plt.show()
print(x)
```

```
plt.subplot(2,1,1)
plt.plot(x,y,color="g",linewidth=3,linestyle="--")
plt.title("First")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.subplot(2,1,2)
plt.plot(x,5*x,color="r",linewidth=3,linestyle="--")
plt.grid(True)
plt.title("Second")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```
student = {"Name":7,"Department":18,"Year":2,"Semester":4}
student1 = list(student.keys())
student2 = list(student.values())
plt.bar(student1,student2,color="r")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My first plot")
plt.xticks(student1,student2,rotation=0)
plt.show()
```
