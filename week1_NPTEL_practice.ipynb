```
n12 = np.vstack((n1,n2))
print(n12)
n21 = np.hstack((n1,n2))
print(n21)
n1122 = np.column_stack((n1,n2))
print(n1122)
```

```
n122 = np.intersect1d(n1,n2)
print(n122)
n133 = np.setdiff1d(n1,n2)
print(n133)
```
