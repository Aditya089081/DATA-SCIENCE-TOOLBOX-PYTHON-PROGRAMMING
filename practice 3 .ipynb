```
import pandas as pd

file_path = "heart_attack_china.csv"

try:
    df = pd.read_csv(file_path, encoding="ISO-8859-1")

    missing_values = df.isnull().sum()
    total_missing = missing_values.sum()

    print("Missing Values per Column:\n", missing_values)
    print("\nTotal Missing Values in Dataset:", total_missing)

    df.fillna(df.median(numeric_only=True), inplace=True)

    categorical_cols = df.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        df[col].fillna(df[col].mode()[0], inplace=True)

    print("\nMissing Values After Processing:\n", df.isnull().sum())

except FileNotFoundError:
    print("Error: File not found!")
    df = pd.DataFrame()
except pd.errors.EmptyDataError:
    print("Error: The file is empty!")
    df = pd.DataFrame()
except pd.errors.ParserError:
    print("Error: The file has inconsistent columns!")
    df = pd.DataFrame()
except UnicodeDecodeError:
    print("Error: Encoding issue, try a different encoding!")
    df = pd.DataFrame()
except Exception as e:
    print(f"Unexpected error: {e}")
    df = pd.DataFrame()

print("\nProcessed DataFrame:\n", df.head())
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "heart_attack_china.csv"

try:
    df = pd.read_csv(file_path, encoding="ISO-8859-1")

    missing_values = df.isnull().sum()
    total_missing = missing_values.sum()

    print("Missing Values per Column:\n", missing_values)
    print("\nTotal Missing Values in Dataset:", total_missing)

    df.fillna(df.median(numeric_only=True), inplace=True)

    categorical_cols = df.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        df[col].fillna(df[col].mode()[0], inplace=True)

    print("\nMissing Values After Processing:\n", df.isnull().sum())

    plt.figure(figsize=(10, 6))
    sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
    plt.title("Missing Values Heatmap")
    plt.show()

    plt.figure(figsize=(8, 5))
    df.hist(figsize=(12, 10), bins=20)
    plt.suptitle("Histograms of Numerical Features")
    plt.show()

    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df.select_dtypes(include=['number']))
    plt.xticks(rotation=90)
    plt.title("Box Plot of Numerical Features")
    plt.show()

    if 'Age' in df.columns and 'Cholesterol' in df.columns:
        plt.figure(figsize=(8, 5))
        sns.scatterplot(x=df['Age'], y=df['Cholesterol'], alpha=0.6)
        plt.title("Age vs Cholesterol")
        plt.xlabel("Age")
        plt.ylabel("Cholesterol")
        plt.show()

except FileNotFoundError:
    print("Error: File not found!")
    df = pd.DataFrame()
except pd.errors.EmptyDataError:
    print("Error: The file is empty!")
    df = pd.DataFrame()
except pd.errors.ParserError:
    print("Error: The file has inconsistent columns!")
    df = pd.DataFrame()
except UnicodeDecodeError:
    print("Error: Encoding issue, try a different encoding!")
    df = pd.DataFrame()
except Exception as e:
    print(f"Unexpected error: {e}")
    df = pd.DataFrame()

print("\nProcessed DataFrame:\n", df.head())
```
