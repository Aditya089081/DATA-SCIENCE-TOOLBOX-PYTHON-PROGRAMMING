```
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```
 #Box plot and Wishers Chart

x = pd.Series([1,2,3,4,5,5,6])
y = 2*x + 1
z = [x,y]
# print(x)
plt.boxplot(x)
plt.show()
```

```
plt.boxplot(x=z)
plt.show()
```

```
#Voilen Plot

plt.violinplot(z)
plt.show()
```

```
#pie chart

store = {"Mango":12,"Orange":34,"Apple":54}
plt.pie(store.values(),labels=store.keys())
plt.show()
```

```
plt.pie(store.values(),labels=store.keys(),colors = ['Yellow','Orange','Red'])
plt.show()
```

```
plt.pie(store.values(),labels=store.keys(),colors = ['Yellow','Orange','Red'],autopct='%0.1f%%')

plt.show()
```

```
#DoughNut Chart
plt.pie(store.values(),labels=store.keys(),radius = 1,autopct = '%0.2f%%')
plt.show()
```

```
plt.pie([1],colors=['w'],radius=0.25)
plt.show()
```
