```
import pandas as pd
import numpy as np
from statistics import mean, median, mode
import matplotlib.pyplot as plt
import seaborn as sns

num = np.array([78,85,90,88,76])
a = mean(num)
print("Mean: ",a)


b = median(num)
print("Median: ",b)


c = mode(num)
print("Mode: ",c)


print("Standard deviation: ",num.std())
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Corrected list lengths
data = {
    'Studey_Hours': [2, 3, 5, 7, 8, 4, 6, 9, 10],
    'Exam_Score': [50, 60, 75, 85, 90, 65, 80, 95, 98]  # Removed extra value
}

df = pd.DataFrame(data)

# Correlation Analysis
print("Correlation Coefficient:", df.corr().iloc[0, 1])

# Scatter plot (Numerical vs Numerical)
plt.figure(figsize=(6, 4))
sns.scatterplot(x=df['Studey_Hours'], y=df['Exam_Score'])
plt.title("Scatter Plot of Study Hours vs Exam Scores")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Corrected data with an added 'Attendance' column
data = {
    'Study_Hours': [2, 3, 5, 7, 8, 4, 6, 9, 10],
    'Attendance': [75, 80, 85, 90, 95, 70, 88, 92, 98],  # Added Attendance values
    'Exam_Score': [50, 60, 75, 85, 90, 65, 80, 95, 98]
}

df = pd.DataFrame(data)

# 3D Scatter Plot (Multivariate Visualization)
fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['Study_Hours'], df['Attendance'], df['Exam_Score'], c='blue', marker='o')

ax.set_xlabel('Study Hours')
ax.set_ylabel('Attendance')
ax.set_zlabel('Exam Scores')

plt.title("3D Scatter Plot of Study Hours, Attendance, and Exam Scores")
plt.show()
```
