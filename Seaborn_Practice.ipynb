```
import seaborn as sns
from matplotlib import pyplot as plt
df = sns.load_dataset("tips")
df.head()
```

```
sns.lineplot(y=df['total_bill'],x=df['sex'],data=df,hue = df['time'])
plt.show()
```

```
sns.lineplot(y=df['total_bill'],x=df['sex'],data=df,hue = df['time'],marker = True)
```

```
sns.lineplot(y=df['total_bill'],x=df['sex'],data=df,hue = df['time'],style="size")
```

```
#Bar plot
sns.set(style="darkgrid")
sns.barplot(x=df['sex'],y=df['total_bill'],data=df)
plt.show()
```

```
sns.countplot(x=df['sex'],data=df)
```
